### Crie um notebook novo e importe os arquivos, transformando-os em DataFrames. Mantenha os campos chave como índices.

```python
import pandas as pd
```

### importando csvs e agrupando

```python
jogos2013 = pd.read_csv('../../../data/jogos_2013.csv')
jogos2014 = pd.read_csv('../../../data/jogos_2014.csv')
jogos2015 = pd.read_csv('../../../data/jogos_2015.csv')
jogos2016 = pd.read_csv('../../../data/jogos_2016.csv')
jogos2017 = pd.read_csv('../../../data/jogos_2017.csv')
```

```python
brasileirao = pd.DataFrame()
brasileirao = pd.concat([jogos2013, jogos2014, jogos2015, jogos2016, jogos2017], axis=0)
```

```python
brasileirao.head()
```

#### mudando index para os ids 

```python
brasileirao.set_index(['id_arena', 'id_clube1', 'id_clube2', 'id_clube_vencedor', 'id_estado_vencedor', 'id_estado_clube1', 'id_estado_clube2'], inplace=True)
```

```python
brasileirao.head()
```

```python
arenas = pd.read_csv('../../../data/arenas.csv')
clubes = pd.read_csv('../../../data/clubes.csv')
estados = pd.read_csv('../../../data/estados.csv')
```

```python
arenas.set_index(['id_arena'], inplace=True)
clubes.set_index(['id_clube'], inplace=True)
estados.set_index(['id_estado'], inplace=True)
```

```python
arenas.head()
```

#### agrupando arenas com campeonato

```python
brasileirao = brasileirao.join(arenas)
```

```python
brasileirao.head()
```

```python
clubes.head()
```

```python
brasileirao = brasileirao.join(clubes, on=['id_clube1'], lsuffix='1', rsuffix='2')
```

```python
brasileirao = brasileirao.join(clubes, on=['id_clube2'], lsuffix='1', rsuffix='2')
```

```python
brasileirao.head()
```

```python
brasileirao = brasileirao.join(estados, on=['id_estado_clube1'], lsuffix='1', rsuffix='2')
```

```python
brasileirao = brasileirao.join(estados, on=['id_estado_clube2'], lsuffix='1', rsuffix='2')
```

```python
brasileirao = brasileirao.join(estados, on=['id_estado_vencedor'])
```

```python
brasileirao.rename(columns={'sigla_estado':'sigla_estado_vencedor'}, inplace=True)
```

```python
brasileirao.rename(columns={'desc_estado':'desc_estado_vencedor'}, inplace=True)
```

```python
brasileirao = brasileirao.join(clubes, on=['id_clube_vencedor'])
```

```python
brasileirao.rename(columns={'desc_clube':'desc_clube_vencedor'}, inplace=True)
```

```python
brasileirao.head()
```

```python
brasileirao.reset_index(drop=True, inplace=True)
brasileirao.head()
```

```python
# biblioteca para visualizar dados ausentes
import missingno as msno
```

```python
msno.bar(brasileirao)
```

#### a. Qual o clube que mais marcou gols em todas as temporadas?

```python
brasileirao.head()
```

```python
brasileirao.groupby('desc_clube1')[['qtd_gols_clube1']].sum().sort_values(by='qtd_gols_clube1', ascending=False).head(1)
```

```python
brasileirao.groupby('desc_clube2')[['qtd_gols_clube2']].sum().sort_values(by='qtd_gols_clube2', ascending=False).head(1)
```

####  Qual o clube que mais marcou gols na temporada de 2015?

```python
brasileirao[brasileirao['data'].str.slice(start=6) == '2015'].groupby('desc_clube1')[['qtd_gols_clube1']].sum().sort_values(by='qtd_gols_clube1', ascending=False).head(1)
```

```python
brasileirao[brasileirao['data'].str.slice(start=6) == '2015'].groupby('desc_clube2')[['qtd_gols_clube2']].sum().sort_values(by='qtd_gols_clube2', ascending=False).head(1)
```

#### Qual foi a maior goleada de todas as temporadas? Maior goleada nesse contexto é igual a maior diferença de gols entre o que um time marcou em relação ao outro.

```python
brasileirao['qtd_gols'] = brasileirao['qtd_gols_clube1'] - brasileirao['qtd_gols_clube2']
```

```python
brasileirao.head()
```

```python
brasileirao.sort_values(by='qtd_gols', ascending=False).head(4)
```

#### e. Qual foi o estado mais vitorioso (quantidade de partidas) em 2016?

```python
brasileirao.groupby('sigla_estado_vencedor')['sigla_estado_vencedor'].count().sort_values(ascending=False)
```
