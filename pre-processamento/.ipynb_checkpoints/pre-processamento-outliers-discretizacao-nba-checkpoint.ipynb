```python
#importando dependencias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importando dados
cafe_manha = pd.read_csv('../../../data/CAFE_DA_MANHA.csv')
```

```python
cafe_manha.head()
```

```python
# criano um novo campo e definindo valor
cafe_manha['qtd']=1
```

### Crie um df pivotado a partir do df original, através do método pandas pivot_table(); 

#### values será a novo campo qtd
#### indice será o codigo da transacao
#### colunas serao os itens

```python
dfCafePivot = cafe_manha.pivot_table(values='qtd',index='Transaction',columns='Item',fill_value=0)
```

```python
dfCafePivot.head()
```

### criando novo df com a equacao de correlação de Pearson sobre a matriz pivotada

```python
corrCafe = dfCafePivot.corr()
```

```python
corrCafe
```

```python
plt.figure(figsize=[22,18])
sns.heatmap(corrCafe,cmap='magma',linecolor='w')
```

### discretizando o campo AGE em um novo campo DESC_Age, de acordo com o quantile. Junior (33%) Médio (66%) Senior (100%)

```python
def dicAge(x):
    if x < dfmjNBA['Age'].quantile(0.333):
        return 'Junior'
    elif x < dfmjNBA['Age'].quantile(0.666):
        return 'Medio'
    else:
        return  'Senior'
```

```python
dfmjNBA['DESC_Age'] = dfmjNBA['Age'].apply(dicAge)
```
