```python
from sklearn import datasets
```

```python
import pandas as pd
```

```python
wineNp = datasets.load_wine()
```

```python
winedf = pd.DataFrame(data=wineNp['data'], columns=wineNp['feature_names'])
winedf.head()
```

```python
winedf['target'] = wineNp['target']
winedf.head()
```

```python
winedf['targetCat'] = winedf['target'].map(lambda ind: wineNp['target_names'][ind])
winedf.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
normWineModel = StandardScaler()
```

```python
normWineModel.fit(winedf.drop(['target','targetCat'], axis=1))
```

```python
normWineData = normWineModel.transform(winedf.drop(['target','targetCat'], axis=1))
```

```python
dfNormWine = pd.DataFrame(normWineData, columns=winedf.drop(['target','targetCat'], axis=1).columns)
```

```python
dfNormWine.head()
```

## Aplicando PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
X_pca = pca.fit_transform(dfNormWine)
```

```python
pca.components_
```

```python
dfWinePCA = pd.DataFrame(data=X_pca, columns=['PCA_1','PCA_2','PCA_3'])
```

```python
dfWinePCA[['target','targetCat']] = winedf[['target','targetCat']]
```
