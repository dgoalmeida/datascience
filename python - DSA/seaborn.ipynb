```python
import seaborn as sns
import pandas as pd
import numpy as np
from  scipy import stats
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.describe()
```

```python
# grafico com estimativa de densidade de kernel - Distreibuição univariada
sns.distplot(iris.sepal_length, rug=True, fit=stats.gausshyper)
```

```python
# grafico  scatterplot - distribuição bivariada
sns.jointplot(x='sepal_length',y='petal_length', data=iris)
```

```python
with sns.axes_style('white'):
    sns.jointplot(x='sepal_length',y='petal_length', data=iris, kind='hex', color='k')
```

```python
sns.jointplot(x='sepal_length',y='petal_length', data=iris, kind='kde')
```

```python
g = sns.jointplot(x='sepal_length',y='petal_length', data=iris, kind='kde', color='m')
g.plot_joint(plt.scatter, c='w',s=30,linewidth=1, marker='+')
g.ax_joint.collections[0].set_alpha(0)
```

```python
sns.pairplot(iris)
```

```python
# Visualizaçãp de relacionamento lienar 
tips = sns.load_dataset('tips')
tips.head()
```

```python
# scatterplot com linha de regreção
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

```python
# regreção linear (utiliza 95% de intervalo de confiança por padrão)
sns.lmplot(x='total_bill',y='tip', data=tips)
```

```python
# adicionando limite superior e inferior
sns.lmplot(x='size',y='tip', data=tips, x_estimator=np.mean)
```

```python
anscombe = sns.load_dataset('anscombe')
anscombe.head()
```

```python
# relacionamento não linear
sns.lmplot(x='x',y='y',data=anscombe.query("dataset == 'II'"), ci = None, scatter_kws={'s':80})
```

```python
sns.lmplot(x='x',y='y',data=anscombe.query("dataset == 'II'"), order=2, ci = None, scatter_kws={'s':80})
```

```python
# usando mais de duas variaveis
sns.lmplot(x='total_bill',y='tip', hue='smoker', data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip', hue='smoker', col='time', data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip', hue='smoker', col='time', row='sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', hue='time', data=tips)
```
