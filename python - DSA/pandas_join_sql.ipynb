## Pandas função SQL JOIN

```python
import pandas as pd
```

```python
dados1 = { 'diciplina_id':[1,2,3,4,5],
           'nome':['Bernardo','Álan','Matheus','Ivo','Gerso'],
           'sobrenome':['Anderson','Teixeira','Amoedo','Trindade','Vargas']}
df_a = pd.DataFrame(dados1, columns=['diciplina_id','nome','sobrenome'])
df_a
```

```python
dados2 = { 'diciplina_id':[4,5,6,7,8],
           'nome':['Roberto','Mariana','Ana','Marcos','Maria'],
           'sobrenome':['Sampaio','Fernandes','Arantes','Amoedo','Martins']}
df_b = pd.DataFrame(dados2, columns=['diciplina_id','nome','sobrenome'])
df_b
```

```python
dados3 = { 'diciplina_id':[1,2,3,4,5,7,8,8,10,11],
           'teste_id':[81,75,75,71,76,84,95,61,57,90]}

df_n = pd.DataFrame(dados3, columns=['diciplina_id','teste_id'])
df_n
```

```python
# Join dos dataframes pelas linhas

df_new = pd.concat([df_a,df_b])
df_new
```

```python
# Join dos dataframes pelas colunas
pd.concat([df_a,df_b], axis=1)
```

```python
# Join de dois dataframes pela coluna diciplina_id
pd.merge(df_new,df_n,on='diciplina_id')
```

```python
# Join de dois dataframes pela coluna diciplina_id
pd.merge(df_new,df_n,left_on='diciplina_id',right_on='diciplina_id')
```

```python
# merge outer join
# A união externa completa produz o conjunto de todos os registros na tabela A e na tabela B
# com registros correspondentes de ambos os lados, quando disponiveis
# se não ouver correspondencia, o lado ausente conterá null
pd.merge(df_a,df_b, on='diciplina_id', how='outer')
```

```python
# Merge inner join
# A unão interna produz apenas o conjunto de registros correspondentes na tabela A e na tabela B
pd.merge(df_a,df_b, on='diciplina_id', how='inner')
```

```python
# Merge left join
# A junção externa esquerda produz um conjunto completo com registros da tabela A
# com os registros correspondentes (quando disponivel) da coluna B
# se não ouver correspondente, o lado direito conterá nulo
pd.merge(df_a,df_b, on='diciplina_id', how='left')
```

```python
# Merge right join o contrario do item anterior
pd.merge(df_a,df_b, on='diciplina_id', how='right')
```

```python
# adicionando sufixo para identificar as colunas
pd.merge(df_a, df_b, on='diciplina_id', how='left', suffixes=['_left','_right'])
```

```python
# Join baseado em indices
pd.merge(df_a,df_b, right_index=True, left_index=True, suffixes=['_left','_right'])
```
