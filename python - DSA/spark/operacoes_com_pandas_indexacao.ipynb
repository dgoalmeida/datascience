## Operações Pandas dataframe indexação


```python
import pandas as pd
import numpy as np
```

```python
dict1 = { 'Coluna1':[190, 231, 784, 127],
         'Coluna2':[545, 278, 104, 347],
         'Coluna3':[123, 456, 789, 763]}
```

```python
df = pd.DataFrame(dict1)
```

```python
df
```

```python
df.dtypes
```

```python
df['Coluna2']
```

```python
df['Coluna2'][0]
```

```python
# inserindo uma coluna na posição 0 com nome ID
df.insert(0,'ID',[1,2,3,4])
```

```python
df
```

```python
# alterando index do data frame
df = pd.DataFrame(df.set_index('ID'))
```

## Manipulando e Organizando dataframes

```python
colunas = list(np.arange(26))
```

```python
import string
```

```python
ucase = [ x for x in string.ascii_uppercase]
lcase = [x for x in string.ascii_lowercase]
```

```python
alfabeto = pd.DataFrame([ucase,lcase,colunas])
```

```python
alfabeto
```

```python
alfabeto = alfabeto.T
```

```python
alfabeto.head()
```

```python
alfabeto.index
```

```python
alfabeto.columns
```

```python
alfabeto.columns = ['Maiuscula','Minuscula','numero']
```

```python
alfabeto
```

```python
# colocando uma data como index
alfabeto.index = pd.date_range('01/01/2021',periods=26)
```

```python
alfabeto
```

```python
#filtrando entre as datas
alfabeto['2021-01-10':'2021-01-19']
```
