utilizando nc (nec cat) para abrir uma tream local
comando nc -lk 9898

```python
from pyspark.streaming import StreamingContext
```

```python
# criando um streamingContect com intervlo batch de 1 segundao
ssc = StreamingContext(sc, 1)
```

```python
# criando um DsStream que vai se conectar na porta 9898 da maquina local
linhas = ssc.socketTextStream('localhost',9898)
```

```python
type(linhas)
```

```python
# dividindo cada linha em palavra
palavra = linhas.flatMap(lambda line: line.split(" "))
```

```python
type(palavra)
```

```python
# conta cada palavra em cada batch
pares = palavra.map(lambda x: (x,1))
contaPalavras = pares.reduceByKey(lambda x,y: x+y)
```

```python
# exibe os 10 primeiros
contaPalavras.pprint()
```

```python
ssc.start() # inicia a coleta e processamento do stream de dados
ssc.awaitTermination() # espera a computação finalizar
```
